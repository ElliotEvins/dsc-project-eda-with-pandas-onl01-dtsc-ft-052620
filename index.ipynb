```python
# Let's get started importing the necessary libraries
import pandas as pd

```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
df.head()
```

```python
# Investigate the Data
df.columns
df.info()
```

```python
# Investigating Distributions using scatter_matrix
# numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

# varlist = df.select_dtypes(include=numerics)
pd.plotting.scatter_matrix(df[['MSSubClass',
 'MSZoning',
 'LotArea',
 'Utilities',
 'Neighborhood',
 'Condition1',
 'Condition2',
 'BldgType',
 'HouseStyle',
 'OverallQual',
 'OverallCond',
 'MasVnrType',
 'MasVnrArea',
 'ExterQual',
 'ExterCond',
 'BsmtExposure',
 'BsmtFinType1',
 'Heating',
 'HeatingQC',
 'CentralAir',
 'Electrical',
 'LowQualFinSF',
 'GrLivArea',
 'BsmtFullBath',
 'BsmtHalfBath',
 'FullBath',
 'HalfBath',
 'BedroomAbvGr',
 'KitchenAbvGr',
 'SaleType',
 'SaleCondition']], figsize=(20,20));
```

```python
# Create a plot that shows the SalesPrice Distribution
df['SalePrice'].plot.hist()
```

```python
# Create a plot that shows the LotArea Distribution
df['LotArea'].plot.hist()
```

```python
# Create a plot that shows the Distribution of the overall house condition
df['OverallCond'].plot.hist()
```

```python
# Create a Box Plot for SalePrice
df['SalePrice'].plot.box(figsize = (8,15))
```

```python
# Perform an Exploration of home values by age
df.plot.scatter('SalePrice', 'YearBuilt', c = 'YearRemodAdd',  colormap = 'viridis');
```
